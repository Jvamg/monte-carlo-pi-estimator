```python
import matplotlib.pyplot as plt
from monte_carlo import forWayBall
from monte_carlo import numpyWayBall
```

```python
n = 100000
```

```python
pi_for, time_for = forWayBall(n)
pi_numpy, time_numpy, x, y, z, w = numpyWayBall(n)
```

```python
mascara = w <= 1
x_dentro = x[mascara]
x_fora = x[~mascara]
y_dentro = y[mascara]
y_fora = y[~mascara]
z_dentro = z[mascara]
z_fora = z[~mascara]
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(x_dentro, y_dentro, z_dentro, color='blue', s=0.1)
ax.scatter(x_fora, y_fora, z_fora, color='red', s=0.1)
ax.set_box_aspect([1, 1, 1])
plt.show()
```

```python
print(f"Pi output from forWay: {pi_for}")
print(f"Pi output from numpyWay: {pi_numpy}")
```

```python
nomes = ['For Way', 'Numpy Way']
tempos = [time_for, time_numpy]
```

```python
plt.bar(nomes, tempos, color=['red', 'green'])
plt.title('Comparing time of execution')
plt.ylabel('Seconds')
plt.show()
print(f"Numpy Way was {(time_for / time_numpy):.1f}x faster than For Way.")
```
