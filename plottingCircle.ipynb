```python
import matplotlib.pyplot as plt
import numpy as np
from monte_carlo import forWayCircle
from monte_carlo import numpyWayCircle
```

```python
n = 100000
```

```python
pi_for, time_for = forWayCircle(n)
pi_numpy, time_numpy, x, y, z = numpyWayCircle(n)
```

```python
mascara = z <= 1
x_dentro = x[mascara]
x_fora = x[~mascara]
y_dentro = y[mascara]
y_fora = y[~mascara]
```

```python
plt.scatter(x_dentro, y_dentro, color='blue', s=1)
plt.scatter(x_fora, y_fora, color='red', s=1)
plt.axis('equal')
plt.show()
```

```python
print(f"Pi output from forWay: {pi_for}")
print(f"Pi output from numpyWay: {pi_numpy}")
```

```python
nomes = ['For Way', 'Numpy Way']
tempos = [time_for, time_numpy]
```

```python
plt.bar(nomes, tempos, color=['red', 'green'])
plt.title('Comparing time of execution')
plt.ylabel('Seconds')
plt.show()
print(f"Numpy Way was {(time_for / time_numpy):.1f}x faster than For Way.")
```
